````python
import re


problematic_columns = []

# additional data issues were found so checking for the columns that have data discrepancy
for column in object_columns:
    try:
        df_Indicators[column] = df_Indicators[column].astype(float)
    except ValueError:
        problematic_columns.append(column)

print("Problematic columns:", problematic_columns)

for column in problematic_columns:
    non_numeric_rows = df_Indicators[df_Indicators[column].apply(lambda x: not re.match(r'^[0-9.]+$', str(x)))]

    print(f"Discrepancy in column '{column}':")
    unique_special_characters = set()
    for index, value in non_numeric_rows[column].items():
        special_characters = re.findall(r'[^0-9.]', str(value))
        unique_special_characters.update(special_characters)
        # print(f"Row {index}: {value}")
        # Exclude specific characters that are considered valid
        # Replace 'a' and 'n' characters with NaN in unique_special_characters set
        unique_special_characters = {char if char not in ['a', 'n'] else np.nan for char in unique_special_characters}

    print("Unique special characters causing errors:", unique_special_characters)
    print()
````

````python
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler
import pandas as pd

df_Cleanwithcountry['Key'] = range(1, len(df_Cleanwithcountry) + 1)
def imputation_using_KNN(df_Cleanwithcountry, selected_columns,imputedcolumn):
    # Standardize the complete data
    # scaler = StandardScaler() # commented since it was actually nullifying the existing values and updating the target variable

    # print(df_Cleanwithcountry[imputedcolumn].isna().sum())
    df_Cleanwithoutcountry = df_Cleanwithcountry.loc[:, df_Cleanwithcountry.columns != 'Country']
    # dfcomplete_scaled = pd.DataFrame(scaler.fit_transform(df_Cleanwithoutcountry), columns=df_Cleanwithoutcountry.columns)

    # Select specific columns from dfcompleteclean with corresponding index values
    df_selected = df_Cleanwithoutcountry[selected_columns].copy()
    print(df_selected[imputedcolumn].isna().sum())
    # Perform KNN imputation on 'Population_Urban'
    imputer = KNNImputer(n_neighbors=5)
    df_selected[imputedcolumn] = imputer.fit_transform(df_selected[[imputedcolumn]])

    # Find the corresponding datapoints in the parent dataframe
    missing_indices = df_Cleanwithoutcountry.index[df_Cleanwithoutcountry[imputedcolumn].isnull()]
    corresponding_values = df_selected.loc[missing_indices, imputedcolumn]

    # Update the corresponding column in the parent dataframe with the imputed values using the Key column
    df_Cleanwithoutcountry.loc[missing_indices, imputedcolumn] = corresponding_values.values

    # Inverse transform the scaled data
    # dfcomplete_imputed = pd.DataFrame(scaler.inverse_transform(dfcomplete_scaled), columns=dfcomplete_scaled.columns)

    # Update the missing values in "imputedcolumn" column of the original dataset
    df_Cleanwithcountry[imputedcolumn] = df_Cleanwithcountry['Key'].map(df_Cleanwithoutcountry.set_index('Key')[imputedcolumn])
    # Save the updated parent dataframe to an Excel file
    output_file = 'imputedcolumn.xlsx'  # Specify the output file path and name
    df_Cleanwithcountry.to_excel(output_file, index=False)
    print('successfully Imputed : ',imputedcolumn,'and saved to Excel')
    return df_Cleanwithcountry


````
